Checking the data quality, looking for nagative values and outliers in each column

```python
#Checking negative values in each column
columns = ['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'Tamb', 'RH', 'WS',
           'WSgust', 'WSstdev', 'WD', 'WDstdev', 'BP', 'Precipitation','TModA', 'TModB']
for column in columns:
    print(f"{column} {bsierraleone_bumbuna_df[bsierraleone_bumbuna_df[column]<0][column].shape}")
```

```python
#Calculating what percentage of the total data has invalid value for "GHI", "DNI" and "DHI" columns 
# in order to figure out the suitable treatment
print((261135/525600)*100)
print((266352/525600)*100)
print((263128/525600)*100)
```

```python
#Replace negative values in each column with numpy NAN
for column_index in [1,2,3]:
    if column_index == 1:
        index_ = bsierraleone_bumbuna_df[bsierraleone_bumbuna_df['GHI']<0]['GHI'].index.values.tolist()
        for j in index_:
            bsierraleone_bumbuna_df.iat[j, 1] = np.NAN
    elif column_index == 2:
        index_ = bsierraleone_bumbuna_df[bsierraleone_bumbuna_df['DNI']<0]['DNI'].index.values.tolist()
        for j in index_:
            bsierraleone_bumbuna_df.iat[j, 2] = np.NAN
    elif column_index == 3:
        index_ = bsierraleone_bumbuna_df[bsierraleone_bumbuna_df['DHI']<0]['DHI'].index.values.tolist()
        for j in index_:
            bsierraleone_bumbuna_df.iat[j, 3] = np.NAN
```

```python
#Checking if the replacement goes well by counting null values
bsierraleone_bumbuna_df.isnull().sum()
```
