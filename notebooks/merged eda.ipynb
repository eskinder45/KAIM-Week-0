```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
binin = pd.read_csv("C:\\Users\\esk\\Desktop\\data\\benin_cleaned.csv")
seralione = pd.read_csv("C:\\Users\\esk\\Desktop\\data\\sierraleone_cleaned.csv")
togo = pd.read_csv("C:\\Users\\esk\\Desktop\\data\\togo_cleaned.csv")
```

```python
print(f"benin: {binin.shape}\nserralione: {seralione.shape}\ntogo: {togo.shape}")
```

```python
# Combine datasets into one for visualization
binin['Region'] = pd.Series(['b' for i in range(525600)])
seralione['Region'] = pd.Series(['s' for i in range(525600)])
togo['Region'] = pd.Series(['t' for i in range(525600)])
```

```python
df = pd.concat([binin, seralione, togo])
```

```python
df['Region'].shape[0]/3
```

```python
# Plot histograms
features = ['GHI', 'DNI', 'DHI']
for feature in features:
    plt.figure(figsize=(10, 6))
    sns.histplot(data=df, x=feature, hue='Region', kde=True, bins=10, palette='muted')
    plt.title(f'Distribution of {feature} Across Regions')
    plt.xlabel(f'{feature} (W/m²)')
    plt.ylabel('Frequency')
    plt.grid(alpha=0.3)
    plt.show()
```

```python
df.set_index('Timestamp', inplace=True)
```

```python
# Bar plot for irradiance variability
plt.figure(figsize=(10, 6))
sns.barplot(x='Region', y='WSstdev', data=df, color='lightblue', label='WSs Std Dev')
sns.barplot(x='Region', y='WDstdev', data=df, color='orange', label='WDS Std Dev', alpha=0.7)
plt.title('Irradiance Variability by Region')
plt.ylabel('Standard Deviation (W/m²)')
plt.xlabel('Region')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
